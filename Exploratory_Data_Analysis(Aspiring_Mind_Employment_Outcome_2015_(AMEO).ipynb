# **Univariate Analysis**

```python
# Plotting the distribution of 'Salary'
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True, color='blue')
plt.title('Salary Distribution')
plt.show()
```

The histogram shows the distribution of salaries among the participants. The curve indicates that the salary distribution is left-skewed, with most salaries clustered around the lower end (below 2 lakhs) and a gradual decline towards higher salaries. The KDE line confirms that a significant number of individuals earn between 1.5 to 2 lakhs.

```python
# Boxplot for Salary (To find outliers)
plt.figure(figsize=(8, 6))
sns.boxplot(df['Salary'], color='orange')
plt.title('Salary Boxplot')
plt.show()
```

The boxplot illustrates the salary distribution, highlighting the median salary at approximately 2 lakhs. Several outliers are visible above the upper whisker, suggesting that a few individuals earn significantly more than the majority. The interquartile range (IQR) indicates that the middle 50% of salaries lie between 1.5 lakhs and 2.5 lakhs.

```python
# Count plot for 'Gender' column
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

```

The countplot reveals that the dataset comprises a balanced gender distribution, with a slight majority of male participants (55%) compared to female participants (45%). This distribution provides a diverse perspective on salary outcomes across genders in the analysis

```python
# Count plot for 'Specialization' column with rotated labels for better readability
plt.figure(figsize=(12, 6))
sns.countplot(x='Specialization', data=df)
plt.title('Specialization Distribution')
plt.xticks(rotation=90)
plt.show()
```

The countplot indicates that 'electronics and communication' is the most common specialization among the participants, followed by 'Computer Science and Engineering' and 'Information Technologues' This distribution suggests a strong interest in technology-related fields, which may correlate with salary levels in the analysis.

```python
# Identifying outliers in the 'Salary' column using Interquartile Range (IQR)
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]
```

# **Bivariate** **Analysis**

```python
# Scatter plot to explore the relationship between '10percentage' and 'Salary'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='10percentage', y='Salary', data=df)
plt.title('Salary vs 10th Percentage')
plt.show()
```

The scatter plot demonstrates a positive correlation between the 10th-grade percentage and salary, suggesting that higher academic performance in early education may lead to better salary outcomes.

```python
# Select only the numerical columns from the dataset for correlation
numerical_columns = df.select_dtypes(include=['float64', 'int64'])  # Select only numerical columns
correlation_matrix = numerical_columns.corr()

# Plot the correlation heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Salary by Gender')
plt.show()
```

The boxplot reveals that male participants have a higher median salary compared to female participants

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Specialization', y='Salary', data=df)
plt.title('Salary by Specialization')
plt.xticks(rotation=90)
plt.show()
```

```python
jobs = ['programmer analyst', 'software engineer', 'hardware engineer', 'associate engineer']
filtered_data = df[df['Designation'].isin(jobs)]
```

```python
fresh_graduate_salary = filtered_data[(filtered_data['Salary'] >= 250000) & (filtered_data['Salary'] <= 300000)]
print(f"Number of fresh graduates earning between 2.5 to 3 lakhs: {len(fresh_graduate_salary)}")
```

```python
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Test p-value: {p}")
```

There is a significant relationship between Gender and Specialization.
